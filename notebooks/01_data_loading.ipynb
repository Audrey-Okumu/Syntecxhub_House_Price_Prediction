```python
from dotenv import load_dotenv
import os
import kaggle
import zipfile
import pandas as pd

load_dotenv()

os.environ["KAGGLE_API_TOKEN"] = os.getenv("KAGGLE_API_TOKEN")
```

```python
kaggle.api.authenticate()
kaggle.api.competitions_list()
```

```python
kaggle.api.competition_download_files(
    'house-prices-advanced-regression-techniques',  # competition name
    path='data',  # folder where you want the files
    quiet=False
)
```

```python
zip_path = 'data/house-prices-advanced-regression-techniques.zip'
extract_path = 'data/'

# Check if unzipped already
if not os.path.exists(os.path.join(extract_path, 'train.csv')):
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(extract_path)
    print("Dataset extracted.")
else:
    print("Dataset already extracted.")
```
